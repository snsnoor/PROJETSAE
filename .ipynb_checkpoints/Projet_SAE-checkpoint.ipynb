```python
# Filtrer les salaires pour un pays donné (ex : US)
ds[ds["company_location"]== "US"]
```

```python
# Top 5 des entreprises qui paient le plus
ds.groupby('company_location')['salary_in_usd'].mean().round(2).sort_values(ascending=False).head(5)
```

```python
# Filtrer les données
# Rôles avec un salaire moyen supérieur à 100,000
salary_mean=ds.groupby("job_title")["salary_in_usd"].mean().round()
salary_mean[salary_mean > 100000]
```

```python
# Pays avec la plus forte proportion de télétravail (remote_ratio = 100)

ds_re = ds[ds['remote_ratio'] == 100]
proportion = (ds_re['company_location'].sort_values(ascending=False)

proportion
```

```python
# Requêtes conditionnelles
# Salaire moyen des employés à distance (100%)
salary_rem= ds[ds['remote_ratio'] == 100]['salary_in_usd'].mean()
print(salary_rem)

# Comparer salaires moyens pour deux catégories
salary_non_rem = ds[ds['remote_ratio'] == 0]['salary_in_usd'].mean()
print(salary_non_rem)   

# On remarque que le salaire moyen de ceux qui font pas de télétravail est plus éléve


```

```python
# Ajouter une colonne pour les fourchettes de salaires

ds['salary_range'] = pd.cut(ds['salary_in_usd'], bins=[0, 50000, 100000, 150000, 200000, np.inf], labels=['<50k', '50k-100k', '100k-150k', '150k-200k', '200k+']) 
ds
```

```python
# Analyse des corrélations
# Relation entre télétravail et salaire
r=np.corrcoef(ds['remote_ratio'],ds['salary_in_usd'])
r
```

Le coefficient de corrélation est négatif et très proche de 0. Les variables télétravail et salaire semble ne pas être correlés

```python
# Salaire moyen par taille d'entreprise
ds.groupby('company_size')['salary_in_usd'].mean().round(2).sort_values(ascending=False)

#Les entreprises de taille M ont un plus salaire moyen élévé, 
#ce qui est un peu bizarre pour moi car je me dis plus lentreprise est grande plus il y a des personnes avec differents salaires
```

```python
# Tableaux croisés
# Moyenne des salaires par niveau d'expérience et télétravail
pivot_table = ds.pivot_table(
    index='experience_level', 
    columns='remote_ratio',
    values='salary_in_usd', 
    aggfunc='mean'
)
print("Tableau croisé des salaires :", round(pivot_table,2))
```

On voit que plus le niveau d’expérience est élevé, plus le salaire augmente, et que le télétravail à 0% est en général associé aux salaires les plus élevés
